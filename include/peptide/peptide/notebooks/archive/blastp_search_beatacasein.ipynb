```python
from Bio import SeqIO
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
from io import StringIO

def calculate_alignment_percentage(fasta_content):
    # Parse FASTA sequences
    sequences = list(SeqIO.parse(StringIO(fasta_content), "fasta"))
    seq1 = str(sequences[0].seq)
    seq2 = str(sequences[1].seq)
    
    # Perform global alignment
    alignments = pairwise2.align.globalxx(seq1, seq2)
    best_alignment = alignments[0]
    
    # Calculate percentage identity
    matches = sum(a == b for a, b in zip(best_alignment.seqA, best_alignment.seqB))
    total_length = len(best_alignment.seqA)
    identity_percentage = (matches / total_length) * 100
    
    # Print alignment details
    print(f"Sequence Length: {total_length}")
    print(f"Matching Positions: {matches}")
    print(f"Alignment Percentage: {identity_percentage:.2f}%")
    print("\nAlignment:")
    print(format_alignment(*best_alignment))
    
    return identity_percentage

# Read the FASTA file content
with open('betacasein.fasta', 'r') as file:
    fasta_content = file.read()

# Calculate and display alignment
percent_identity = calculate_alignment_percentage(fasta_content)
```
